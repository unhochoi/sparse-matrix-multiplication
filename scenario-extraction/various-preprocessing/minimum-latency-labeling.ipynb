```python
import pandas as pd
import numpy as np

df = pd.read_csv('./csv/spmm-square-nonsquare-data-over-3s-doptimal-878.csv')
df
```

```python
min_label = []

for i in range(len(df)):
    latency_list = []
    latency_list.append(df.loc[i,'sp_smdm'])
    latency_list.append(df.loc[i,'bz_smsm'])
    latency_list.append(df.loc[i,'bz_smdm'])
    min_label.append(latency_list.index(min(latency_list)))

df = pd.concat([df,pd.DataFrame(min_label,columns=['min_label'])], axis =1)
df.to_csv('./csv/spmm-square-nonsquare-data-over-3s-doptimal-878-labeling.csv', index=False)   # csv 저장시 index 제거
df
```
