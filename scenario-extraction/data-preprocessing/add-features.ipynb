```python
import pandas as pd
import numpy as np

df = pd.read_csv('/Users/bdlab/Desktop/sparse-matrix-multiplication/scenario-extraction/d-optimal/spmm-of-d-optimal-data/lr-150000-lc-100000-rc-50000/1293-spmm-over-3s-without-add-feature.csv')  
df
```

```python
feature = df[['lr','lc','rc','ld','rd','lnnz','rnnz']]
target = df[['sp_smdm','bz_smsm','bz_smdm']]
```

```python
import math

lrlc_list=[]
lcrc_list=[]
lrrc_list=[]
lrlcrc_list=[]
ldrd_list=[]
lrrcldrd_list=[]
lrlcrcldrd_list=[]

for i in range(len(df)):
    lrlc_list.append(feature.iloc[i,0] * feature.iloc[i,1])
    lcrc_list.append(feature.iloc[i,1] * feature.iloc[i,2])
    lrrc_list.append(feature.iloc[i,0] * feature.iloc[i,2])
    lrlcrc_list.append(feature.iloc[i,0] * feature.iloc[i,1] * feature.iloc[i,2])
    ldrd_list.append(feature.iloc[i,3] * feature.iloc[i,4])
    lrrcldrd_list.append(feature.iloc[i,0] * feature.iloc[i,2] * feature.iloc[i,3] * feature.iloc[i,4])
    lrlcrcldrd_list.append(feature.iloc[i,0] * feature.iloc[i,1] * feature.iloc[i,2] * feature.iloc[i,3] * feature.iloc[i,4]) 
```

```python
lrlc_df = pd.DataFrame(lrlc_list,columns=['lr*lc'])
lcrc_df = pd.DataFrame(lcrc_list,columns=['lc*rc'])
lrrc_df = pd.DataFrame(lrrc_list,columns=['lr*rc'])
lrlcrc_df = pd.DataFrame(lrlcrc_list,columns=['lr*lc*rc'])
ldrd_df = pd.DataFrame(ldrd_list,columns=['ld*rd'])
lrrcldrd_df = pd.DataFrame(lrrcldrd_list,columns=['lr*rc*ld*rd'])
lrlcrcldrd_df = pd.DataFrame(lrlcrcldrd_list,columns=['lr*lc*rc*ld*rd'])

add_feature = pd.concat([lrlc_df,lcrc_df,lrrc_df,lrlcrc_df,ldrd_df,lrrcldrd_df,lrlcrcldrd_df],axis=1) 
add_feature
```

```python
total_feature = pd.concat([feature, add_feature],axis=1) 
total_feature
```

```python
total_feature.to_csv('/Users/bdlab/Desktop/sparse-matrix-multiplication/scenario-extraction/d-optimal/spmm-of-d-optimal-data/lr-150000-lc-100000-rc-50000/1293-spmm-over-3s.csv', index=False)   # csv 저장시 index 제거
```
