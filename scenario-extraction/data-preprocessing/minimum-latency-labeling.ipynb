```python
import pandas as pd
import numpy as np

df = pd.read_csv('/Users/bdlab/Desktop/sparse-matrix-multiplication/scenario-extraction/d-optimal/spmm-of-d-optimal/lr-150000-lc-100000-rc-50000/1293-spmm-over-3s.csv')  
df
```

### 최소 latency label 확인

```python
min_label = []

for i in range(len(df)):
    latency_list = []
    latency_list.append(df.loc[i,'sp_smdm'])
    latency_list.append(df.loc[i,'bz_smsm'])
    latency_list.append(df.loc[i,'bz_smdm'])
    min_label.append(latency_list.index(min(latency_list)))
```

### 최소  latency label 결과 추가

```python
df = pd.concat([df,pd.DataFrame(min_label,columns=['min_label'])], axis =1)
df.to_csv('/Users/bdlab/Desktop/sparse-matrix-multiplication/scenario-extraction/d-optimal/spmm-of-d-optimal/lr-150000-lc-100000-rc-50000/1293-spmm-over-3s-with-label.csv', index=False)   # csv 저장시 index 제거
df
```
